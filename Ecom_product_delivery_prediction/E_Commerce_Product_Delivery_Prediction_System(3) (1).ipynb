        <strong style="color: #1ABC9C;">• What is a Correlation Heatmap? •</strong>
        <br><br>
        A correlation heatmap is a visual representation of the correlation coefficients between various numerical variables in a dataset. Each cell of the heatmap represents the correlation between two variables, with values ranging from -1 to +1. This allows us to quickly understand the relationships between different features in the dataset.
        <br><br>
        <strong style="color: #1ABC9C;">• Types of Correlation Coefficients •</strong>
        <br><br>
        <strong>- Strong Positive Correlations (near +1):</strong>
        <br>
        When two variables have a high positive correlation, it means that as one variable increases, the other tends to increase as well. For example, in a sales dataset, the number of advertisements run and the number of units sold might show a strong positive correlation.
        <br><br>
        <strong>- Strong Negative Correlations (near -1):</strong>
        <br>
        A strong negative correlation indicates an inverse relationship. When one variable increases, the other tends to decrease. For instance, the number of hours spent on social media might have a negative correlation with academic performance.
        <br><br>
        <strong>- Weak or No Correlations (around 0):</strong>
        <br>
        Variables that have a correlation coefficient near zero indicate little to no linear relationship. These variables are independent of each other, meaning that changes in one do not significantly affect the other. An example might be the correlation between shoe size and test scores.
        <br><br>
        <strong style="color: #1ABC9C;">• Why is Correlation Important in Data Analysis? •</strong>
        <br><br>
        Understanding correlations is essential for making informed decisions in data analysis. Here’s how it can help:
        <ul style="margin-top: 10px; padding-left: 20px;color:white;">
            <li><strong>Feature Selection:</strong> Correlation analysis helps identify redundant features. If two features are highly correlated, one can be dropped to reduce multicollinearity.</li>
            <li><strong>Model Improvement:</strong> Understanding variable relationships helps choose features that enhance model performance and prediction accuracy.</li>
            <li><strong>Data Insights:</strong> Correlation analysis provides insights into the underlying trends and relationships within the data, which can guide business decisions.</li>
        </ul>
        <br><br>
        <font color="white">By analyzing the heatmap, we can gain valuable insights that improve our models and lead to more accurate predictions.
    </p>

```python
def visualize_after_remove_outlier(col_name, data):
    # Sanity Check
    fig, axes = plt.subplots(2, 2, figsize = (15,9))

    sns.boxplot(ax = axes[0,0], y = col_name, data = df, color='teal')
    axes[0,0].set_ylabel(ylabel = col_name, fontsize = 12)
    axes[0,0].set_title(col_name+' Before Remove Outlier', fontdict={'fontsize':14, 'fontweight':'bold'})

    sns.boxplot(ax = axes[0,1], y = col_name, data = data, color='teal')
    axes[0,1].set_ylabel(ylabel = col_name, fontsize = 12)
    axes[0,1].set_title(col_name+ ' After Remove Outlier', fontdict={'fontsize':14, 'fontweight':'bold'})

    # Distibution Before Remove Outlier
    sns.distplot(ax = axes[1,0], x = df[col_name], color= 'red')
    axes[1,0].set_xlabel(xlabel = col_name, fontsize = 12)
    axes[1,0].set_ylabel(ylabel = 'Density', fontsize = 12)
    axes[1,0].set_title(col_name+' Before Remove Outlier', fontdict={'fontsize':14, 'fontweight':'bold'})

    # Distribution After Remove Outlier
    sns.distplot(ax = axes[1,1], x = data[col_name], color= 'seagreen')
    axes[1,1].set_xlabel(xlabel = col_name, fontsize= 12)
    axes[1,1].set_ylabel(ylabel = 'Density', fontsize = 12)
    axes[1,1].set_title(col_name+' After Remove Outlier', fontdict={'fontsize':14, 'fontweight':'bold'})
    plt.tight_layout()
```

```python
# IQR method for removing outliers
import numpy as np

def remove_outliers_iqr(df, column_name):
    Q1 = df["Discount_offered"].quantile(0.25)
    Q3 = df["Discount_offered"].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df["Discount_offered"] >= lower_bound) & (df["Discount_offered"] <= upper_bound)]

# Apply IQR method for all numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns
for column in numerical_columns:
    df = remove_outliers_iqr(df, column)

print("Outliers removed using the IQR method.")
```

    <p style="background-color: #2980b9; color: white; font-size: 2.5em; text-align: left; padding: 10px; border-radius: 10px 10px 0 0;">
        • Summary: Outlier Removal Function
    </p>
    <p style="color: #ecf0f1; font-size: 1.4em; line-height: 1.6;">
        <strong>Function Definition:</strong>
        <code style="background-color: #34495e; padding: 5px 10px; border-radius: 5px; color: #ecf0f1; font-size: 1.2em;">
            remove_outliers(df, column)
        </code>
        is a function designed to remove outliers from a specified column in a DataFrame <code>df</code>. It performs the following steps:
    </p>
    <ul style="color: #ecf0f1; font-size: 1.4em; line-height: 1.6;">
        <li><strong>Step 1:</strong> Calculates the first quartile (Q1) and third quartile (Q3) of the specified column.</li>
        <li><strong>Step 2:</strong> Computes the Interquartile Range (IQR), the difference between Q3 and Q1.</li>
        <li><strong>Step 3:</strong> Defines lower and upper bounds to identify potential outliers.</li>
        <li><strong>Step 4:</strong> Filters the DataFrame to include only the rows where the column values fall within these bounds.</li>
    </ul>
    <p style="color: #ecf0f1; font-size: 1.4em; line-height: 1.6;">
        <strong>Purpose:</strong> The goal of this function is to clean the data by removing outliers that could skew the analysis or model performance.
    </p>
    <p style="color: #ecf0f1; font-size: 1.4em; line-height: 1.6;">
        Outliers are defined as values falling outside 1.5 times the IQR from the first and third quartiles. These extreme values are removed from the dataset to improve data quality and ensure more reliable results from data analysis and machine learning models.

```python
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    report = classification_report(y_test, y_pred, output_dict=True)

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

    # Extracting metrics
    metrics = {
        "precision_0": report["0"]["precision"],
        "precision_1": report["1"]["precision"],
        "recall_0": report["0"]["recall"],
        "recall_1": report["1"]["recall"],
        "f1_0": report["0"]["f1-score"],
        "f1_1": report["1"]["f1-score"],
        "macro_avg_precision": report["macro avg"]["precision"],
        "macro_avg_recall": report["macro avg"]["recall"],
        "macro_avg_f1": report["macro avg"]["f1-score"],
        "accuracy": accuracy_score(y_test, y_pred)
    }

    # Convert dictionary to dataframe
    df = pd.DataFrame(metrics, index=[model_name]).round(2)

    return df
```
